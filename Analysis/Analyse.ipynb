# Introduction
In this analysis, we aim to study how generated rotated ligands distribute in the space of interest pre-minimization. The function __angles_generator__ supports two different distributions, `uniform` and `normal`. For each, 500, 1000, 5000, 10000, and 20000 ligands' conformations were randomly generated. Then, we plot their RMSD value to the true conformation. 

Complex and random seed used are `1AY7` and `distribution.seed`, respectively.

### Notice

```python
import os
import numpy as np
import math
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Change working directory to the root
os.chdir('../')
```

# Plot

Graphic presentation using `seaborn`

```python
df = pd.read_csv('Analysis/Results/analyse_result.txt',sep='\t')
```

```python
df
```

```python
# Plot RMSD histogram, the left column is uniform distribution and the right one is normal distribution
CPX = '1PPJ_CP'
g = sns.FacetGrid(df[(df.Minimizer==0) & (df.Cpx==CPX)],hue='Angle')
g = g.map(sns.distplot, "RMSD").add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('RMSD distribution before minimization of {}'.format(CPX))
g.savefig('Angle_before_{}.png'.format(CPX))
plt.show()
```

```python
# Plot RMSD after minimization for a given complex
CPX = '1PPJ_CP'
g = sns.FacetGrid(df[(df.Minimizer==1) & (df.Cpx==CPX)],hue='Angle')
g = g.map(sns.distplot, "RMSD").add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('RMSD distribution after minimization of {}'.format(CPX))
g.savefig('Angle_after_{}.png'.format(CPX))
plt.show()
```
